### ***1. Reading Real Time Stock Price Data***

Imagine having real-time stock price data from industry-favorite sources like Yahoo Finance and Google Finance at your fingertips. With Python, you can easily access stock prices, financial news, and market trends in real-time. The Yahoo Finance library allows you to fetch real-time stock prices, financial statements, and other relevant data from Yahoo Finance. Google Finance also provides an Application Programmming Interface (API) that can be used to fetch real-time financial data. With Python, you can track the stock market and make informed investment decisions in real-time.

The sample code below is used to collect real-time stock price data from Yahoo Finance for Apple, Microsoft and Google and display them in a tabular format. For simplicity of exposition, only the opening and closing prices are shown for the year of 2016.

```python
# ref: https://www.learndatasci.com/tutorials/python-finance-part-yahoo-finance-api-pandas-matplotlib/
!pip install yfinance
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yfin
import matplotlib.pyplot as plt

yfin.pdr_override()
start = '2016-01-01'
end = '2023-12-31'

stocks = ['AAPL','MSFT','GOOG']

df = pdr.get_data_yahoo(stocks, start, end)
lowHigh = df[['Open','Close']]
lowHigh = lowHigh.fillna(method='ffill')
lowHigh.head()
```

### ***2. Using Time Series Data and Data Visualization***

Python is a game-changer for analyzing time series data in finance. With its powerful libraries and modules, Python makes it easy to extract meaningful statistics, forecast data trends, and predict future stock prices. Plus, its data visualization capabilities with libraries like Pandas and Matplotlib allow investors to spot profitable patterns and make informed decisions. Say goodbye to boring spreadsheets, and hello to dynamic data visualization with Python!

The example code below was used to plot the MSFT time-series. It shows how the stock behaves compared to a short and longer term moving average of the price.

```python
close = df['Close']
close = close.fillna(method='ffill')

# Get the MSFT timeseries. This now returns a Pandas Series object indexed by date.
msft = close.loc[:, 'MSFT']

# Calculate the 20 and 100 days moving averages of the closing prices
short_rolling_msft = msft.rolling(window=20).mean()
long_rolling_msft = msft.rolling(window=100).mean()

# Plot everything by leveraging the very powerful matplotlib package
fig, ax = plt.subplots(figsize=(16,9))

ax.plot(msft.index, msft, label='MSFT')
ax.plot(short_rolling_msft.index, short_rolling_msft, label='20 days rolling')
ax.plot(long_rolling_msft.index, long_rolling_msft, label='100 days rolling')

ax.set_xlabel('Date')
ax.set_ylabel('Adjusted closing price ($)')
ax.legend()
```
