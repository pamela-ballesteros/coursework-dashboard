## **Reading Excel files**
3. To read an existing Excel file containing your stock data, you can use the openpyxl.load_workbook() function. For example, if your file is named "stocks.xlsx" and is located in the same directory as your jupyter notebook, you can use the following code to open it. The contents of the file is given below for reference.

```python
from google.colab import drive
drive.mount("/content/drive", force_remount=True)
workbook = openpyxl.load_workbook('/content/drive/MyDrive/stocks.xlsx')
```

Screenshot 2023-01-24 at 2.42.02 PM.png

4. Once the file is open, you can access the different sheets within the workbook using the workbook.worksheets attribute. You can loop through the sheets, for example:

```python
for sheet in workbook:
    print(sheet.title)
```

5. To access the data within a specific sheet, you can use the sheet's cell properties. For example, to access the value of cell A1 in the first sheet, you can use the following code:

```python
sheet = workbook.worksheets[0]
print(sheet['A2'].value)
```

6. You can also loop through the rows and columns of a sheet to access all the data. The following code will print the values of all cells in the first sheet:

```python
for row in sheet.iter_rows():
    for cell in row:
        print(cell.value, end=" ")
    print()
```

## **Creating Excel files**

7. To create a new Excel file, you can use the openpyxl.Workbook() function. For example:

```python
new_workbook = openpyxl.Workbook()
```

8. Once you have created the new workbook, you can create new sheets within it by using the workbook.create_sheet() function. For example:

```python
new_sheet = new_workbook.create_sheet("new_sheet")
```

9. You can then add data to the sheet by accessing the cells and setting their values. For example:

```python
new_sheet['A1'] = 'Ticker'
new_sheet['B1'] = 'Price'
new_sheet['A2'] = 'AAPL'
new_sheet['B2'] = 100
```

10. Once you have finished adding data to the sheet, you can save the workbook to an Excel file using the workbook.save() function. For example:

```python
new_workbook.save('new_stocks.xlsx')
```

A new file will appear on the Files section on the left.



We can read the new Excel file to check its contents (refer to step 6).

```python
new_new_workbook = openpyxl.load_workbook('new_stocks.xlsx')
new_new_sheet = new_new_workbook.get_sheet_by_name('new_sheet')

for row in new_new_sheet.iter_rows():
    for cell in row:
        print(cell.value, end=" ")
    print()
```

## **Modifying and deleting cells**
11. To modify the value of a cell, you can simply access the cell and set its value. For example:

```python
new_new_workbook = openpyxl.load_workbook('new_stocks.xlsx')
new_new_sheet = new_new_workbook.get_sheet_by_name('new_sheet')

new_new_sheet['B2'] = 110

for row in new_new_sheet.iter_rows():
    for cell in row:
        print(cell.value, end=" ")
    print()
```

12. To delete a cell, you can use the sheet.delete_rows() or sheet.delete_cols() method. For example, to delete the second row of the sheet, you can use the following code:

```python
new_new_workbook = openpyxl.load_workbook('new_stocks.xlsx')
new_new_sheet = new_new_workbook.get_sheet_by_name('new_sheet')

new_new_sheet.delete_rows(2)

for row in new_new_sheet.iter_rows():
    for cell in row:
        print(cell.value, end=" ")
    print()
```

13. To delete a column, you can use the sheet.delete_cols() method. For example, to delete the second column of the sheet, you can use the following code:

```python
new_new_workbook = openpyxl.load_workbook('new_stocks.xlsx')
new_new_sheet = new_new_workbook.get_sheet_by_name('new_sheet')

new_new_sheet.delete_cols(2)

for row in new_new_sheet.iter_rows():
    for cell in row:
        print(cell.value, end=" ")
    print()
```

## **Converting Excel files to pandas dataframes**

14. The openpyxl module can also be used in conjunction with pandas to convert Excel files to pandas dataframes. To do this, you will first need to install the pandas module by running the following command in your terminal:

```python
!pip install pandas
```

15. Next, you will need to import the pandas module into your Python script:

```python
import pandas as pd
```

16. To convert an Excel file to a pandas dataframe, you can use the pandas.read_excel() function. For example, if you have an Excel file named "stocks.xlsx" in the same directory as jupyter notebook, you can use the following code to convert it to a dataframe:

```python
df = pd.read_excel('/content/drive/MyDrive/stocks.xlsx')
```

17. You can also specify the sheet name in case you have multiple sheets in your excel file.

```python
df = pd.read_excel('/content/drive/MyDrive/stocks.xlsx', sheet_name='NIO', index_col="Date")

df.head()
```

18. Once the file is converted to a dataframe, you can perform various operations on it, such as filtering, sorting, and calculating statistics.

```python
import matplotlib.pyplot as plt
import seaborn as sns

#create variable M to represent months and calculate mean 
data_monthly = df["Close"].resample('M').mean() 

# Print closing mean for 12 months
data_monthly = data_monthly.to_frame()

# visualizing the data with matplotlib and seaborn libraries (more details on how to use them in tutorial 2)
sns.barplot(x=data_monthly.index, y=data_monthly['Close'])
plt.xticks(rotation=90)
```
