## **Google and Yahoo Finance APIs**
---
There are a few different libraries in Python that can be used to collect stock data from Google Finance or Yahoo Finance. One popular library is yfinance, which can be used to download historical stock data from Yahoo Finance. Another library is pandas_datareader, which can be used to download stock data from a variety of sources, including Google Finance and Yahoo Finance. Here is an example of how to use the yfinance library to download historical stock data for a particular stock:

```python
!pip install yfinance
```

```python
import yfinance as yf

# Define the stock symbol
stock_symbol = "AAPL"

# Download historical stock data
stock_data = yf.download(stock_symbol)

# Print the stock data
stock_data.head()
```

This code will download historical stock data for Apple Inc. (AAPL) and print the data to the console. The data will include information such as the stock's open and close prices, volume, and other relevant data. You can also use the yfinance library to download stock data for multiple stocks at once by passing a list of stock symbols to the download() function. You can also specify a date range for the data that you want to download. For example, to download data for the past year you can use the following code:

```python
import datetime

# Define the stock symbol
stock_symbol = "AAPL"

# Define the date range
start_date = "2020-01-27"
end_date = "2021-01-27"

# Download historical stock data
stock_data = yf.download(stock_symbol, start=start_date, end=end_date)

# Print the stock data
stock_data.head()
```

This will download the stock data for the past year. You should also keep in mind that both google and yahoo finance has discontinued their APIs, so you will have to use other alternatives like Alpha Vantage, IEX Cloud or any other paid APIs to get stock data

## **Alpha Vantage API**
---
One popular alternative is the Alpha Vantage API, which provides stock data including historical data, real-time data, and financial data.

Here is a step by step guide on how to use the Alpha Vantage API to collect stock data in Python:

1. Go to the Alpha Vantage website (https://www.alphavantage.co/) and sign up for a free API key.

2. Install the `alpha_vantage` library in Python using the following command:

```python
!pip install alpha_vantage
```

3. Import the `alpha_vantage` library and the other necessary libraries in your Python script:

```python
from alpha_vantage.timeseries import TimeSeries
import matplotlib.pyplot as plt
```

4. Initialize the TimeSeries object by passing in your API key. You can use this course generated API key if you do not feel comfortable to create one. We also set the `output_format` parameter to `pandas` so that the data is returned in the form of a pandas DataFrame object.

```python
ts = TimeSeries(key='TNYRBUGXHVRQFWZ9', output_format='pandas')
```

5. Use the `get_daily()` or `get_daily_adjusted()` function of the TimeSeries object to retrieve stock data by passing the stock symbol. The `get_daily()` function returns the unadjusted daily data, while the `get_daily_adjusted()` function returns the adjusted daily data:

```python
data, meta_data = ts.get_daily_adjusted('MSFT')
```

6. Use the pandas library to manipulate and analyze the stock data. For example, you can use the head() function to display the first few rows of the data:

```python
import pandas as pd
df = pd.DataFrame(data)
df.head()
```

7. You can also use the matplotlib library to visualize the data. For example, you can use the plot() function to create a line chart of the stock data:

```python
df.plot()
plt.title('Stock data for MSFT')
plt.show()
```

8. You can also use the get_intraday() function of the TimeSeries object to retrieve intraday data by passing the stock symbol and the interval of the data.

```python
data, meta_data = ts.get_intraday('MSFT', interval='1min', outputsize='full')

df = pd.DataFrame(data)
df.head()
```

```python
df.plot()
plt.title('Stock data for MSFT')
plt.show()
```
