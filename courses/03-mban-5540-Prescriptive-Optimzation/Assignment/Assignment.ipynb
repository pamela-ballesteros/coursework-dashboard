# Assignment Problem
\begin{align*}
\min \quad & \sum_{i=1}^{m} \sum_{j=1}^{n} c_{ij}x_{ij} & \\
\text{S.t.:} \quad & \sum_{i=1}^{m} x_{ij} = 1 & (j = 1, 2, \dots, n) \\
& \sum_{j=1}^{n} x_{ij} \leq 1 & (i = 1, 2, \dots, m) \\
& x_{ij} \in \{0,1\} & (i = 1, 2, \dots, m,\ j = 1, 2, \dots, n)
\end{align*}

### Parameters
- $m$ and $n$ are the number of people and jobs respectively ($m \geq n$).
- $c_{ij}$ is the cost of assigning person $i$ to job $j$.

### Decision Variables
- $x_{ij} =  
\begin{cases}
    1,& \text{if person} \ i \ \text{is assigned to job} \ j\\
    0,              & \text{otherwise}
\end{cases}$
