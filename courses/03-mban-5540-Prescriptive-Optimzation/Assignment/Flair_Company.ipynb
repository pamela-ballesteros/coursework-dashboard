### Problem Description:

Flair Furniture Company produces tables and chairs:
- Each table produced results in \\$7 profit, each chair produced results in \\$5 profit.
- Each table requires: 3 hours of carpentry and 2 hours of painting.
- Each chair requires: 4 hours of carpentry and 1 hour of painting.
- Available production capacity: 2,400 hours of carpentry time and 1,000 hours of painting time.
- The marketing department wants Flair to make no more than 450 chairs.
- The marketing department wants Flair to make at least 100 tables.

Determine best possible combination of tables and chairs to manufacture in order to attain maximum
profit.


### Problem Formulation:
Then, the formulation is the following:
$$
\begin{align}
    \max \quad & 7T + 5C \\
    \text{s.t.} \quad & 3T + 4C \leq 2,400 \\
    & 2T + C \leq 1,000 \\
    &  C \leq 450 \\
    & T \geq 100 \\
    & T,C \in \mathbb{R}^+
\end{align}
$$

### Define decision variables:

By default, Gurobi assumes variables are continuous and non-negative ( $\ge 0$ ).

`addVar ( lb=0.0, ub=float('inf'), obj=0.0, vtype=GRB.CONTINUOUS, name="", column=None )`

Arguments:

- lb (optional): Lower bound for new variable.

- ub (optional): Upper bound for new variable.

- obj (optional): Objective coefficient for new variable.

- vtype (optional): Variable type for new variable (GRB.CONTINUOUS, GRB.BINARY, GRB.INTEGER, GRB.SEMICONT, or GRB.SEMIINT).


### Define the constraints:
We define constraints using the `m.addConstr(constr, name="")` function


### Define Objective Function
The `setObjective()` method is used to set the Objective function. We need to provide two things: the mathematical expression for objective and the "sense" (whether we want to maximize or minimize).


### Update and Write the model
To ensure Gurobi has processed all the variables, constraints, and the objective function we just added, we should update the model using `update()` to synchronize everything. Afterward, it is a great practice to write the model to a file usin `write("Name.lp)` so we can see the underlying math in a readable format.

###Solve the model
We use `optimize()` to run the Gurobi engine.

###Checking the Soltion
It is always recommended to check if the model is solved to optimality using `Status()` method.

- `GRB.OPTIMAL`	The best possible solution was found.
- `GRB.INFEASIBLE`	Your constraints are too strict; no solution exists that satisfies all of them.
- `GRB.UNBOUNDED`	You forgot a constraint (like a limit on production), so the objective can grow to infinity.
- `GRB.INF_OR_UNBD`	The model is either infeasible or unbounded.

###Extracting the Solution
- `getVars()` method retrieves a list of all variables in the model.

- `.x` is an attribute that holds the optimal value.

- `.VarName` is an attribute that stores the string name you assigned to the variable.

- `ObjVal` is the attribute that stores the final value of your objective function.
