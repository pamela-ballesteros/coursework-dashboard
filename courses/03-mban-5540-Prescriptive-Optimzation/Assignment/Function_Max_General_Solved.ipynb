###Objective:
We want to write a general function that creates and solves LP models in the  below form:

$$
\begin{align}
Max \ cx \\
s.t. Ax & \leq b \\
x & \geq 0 \\
\end{align}
$$

### Import other needed Libraries here

```python
import numpy as np
import time
```

Set the number of decision variables(n) and constraints (m)

```python
n = 100 # Number of Decision Variables
m = 80 # Nimber of Constraints
```

Declare the matrix and vectors we need, i.e., $A$, $b$, $c$.

```python
c = np.random.randint(1, 101, n)
A = np.random.randint(1, 10, size = (m,n))
b = np.random.randint(1000, 5000, m)
```



### Declare the Function

```python
def Max_Vector(A, b, c):
    model = gp.Model('max_vector')
    x = model.addMVar(len(c), vtype = GRB.INTEGER, name = "x")
    csr = model.addConstr( A @ x <= b)
    model.setObjective( c @ x , GRB.MAXIMIZE)
    model.update()
    model.write('Func_Max_Genral.lp')
    return model

```

###Call the funcion

```python
model = Max_Vector(A,b,c)
```

### Solve the model

```python
start = time.time()
model.optimize()
end = time.time()
```

```python
print("Optimal Value:", model.objVal)
for v in model.getVars():
    if v.x > 1e-6:
        print(v.VarName, v.x)

print("Time to solve:", end - start, "seconds")
```

```python
def General_Vector(A, b, c, mtype, sign):
    model = gp.Model('general_vector')
    x = model.addMVar(len(c), vtype = "I", name = "x")

    if sign == "<=":
        csr = model.addConstr( A @ x <= b)
    else:
        csr = model.addConstr( A @ x >= b)

    model.setObjective( c @ x , mtype)

    model.update()
    model.write('Func_Genral.lp')
    return model
```

```python
modelG = General_Vector(A, b, c, GRB.MINIMIZE, ">=")
```

```python
start = time.time()
modelG.optimize()
end = time.time()
```

```python
print("Optimal Value:", modelG.objVal)
for v in modelG.getVars():
    if v.x > 1e-6:
        print(v.VarName, v.x)

print("Time to solve:", end - start, "seconds")
```
