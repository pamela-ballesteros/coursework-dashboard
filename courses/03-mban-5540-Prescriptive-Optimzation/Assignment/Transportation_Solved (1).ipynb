# Transportation Problem

\begin{align*}
\text{Minimize} \quad & \sum_{i=1}^{m} \sum_{j=1}^{n} c_{ij}x_{ij} & \\
\text{Subject to} \quad & \sum_{j=1}^{n} x_{ij} \leq S_i & (i = 1, 2, \dots, m) \\
& \sum_{i=1}^{m} x_{ij} \geq D_j & (j = 1, 2, \dots, n) \\
& x_{ij} \geq 0 & (i = 1, 2, \dots, m, j = 1, 2, \dots, n)
\end{align*}


### Parameters
- $m$ and $n$ are the number of supply and demand locations respectively.
- $c_{ij}$ is the cost of shipping one unit of goods from supply location $i$ to demand location $j$.
- $S_i$ is the number of units of goods that can be shipped from supply location $i$.
- $D_j$ is the number of units of goods that need to be shippedto demand location $j$.

### Decision Variables
- $x_{ij}$: the number of units of goods shipped from supply location $i$ to demand location $j$.

```python
import numpy as np
import time
```

###2. Parameter Settings

Random generation seed

```python
rnd = np.random
#rnd.seed(0)
```

Then, we can randomly generate:
- the number of origins
- the number of  destinations

We also create a list for each.


```python
o = 20
d = 20
```

```python
O = [i for i in range(1, o + 1)]
D = [j for j in range(1, d + 1)]
```

Adding arcs between all origins/destinations

```python
A = [(i, j) for i in O for j in D]
```

Transportation cost for each arc

```python
c = {(i, j): rnd.randint(1, 100) for i, j in A}
```

Randomly generating supplies and demands for all nodes

```python
Supply = {i: rnd.randint(10, 1000) for i in O}
Demand = {i: rnd.randint(10, 1000) for i in D}
```

Making sure that Total Supply >= Total Demand

```python
if sum(Supply.values()) < sum(Demand.values()):
    diff = sum(Demand.values()) - sum(Supply.values())
    Supply[d] += diff
```

###4. Model Preparation

Model Initialization


```python
model = gp.Model('Transportation')
```

Adding decision variables

```python
x = model.addVars(O, D, vtype=GRB.CONTINUOUS, name="x")
```

Adding supply constraints

```python
for i in O:
    model.addConstr(quicksum(x[i, j] for j in D ) <= Supply[i] , name="Supply(%s)" % i)
```

Adding demand constraints

```python
model.addConstrs((quicksum(x[i, j] for i in O ) >= Demand[j] for j in D), name="Demand")
```

Adding the objective function:

```python
model.setObjective(quicksum(x[i, j]*c[i, j] for i, j in A), GRB.MINIMIZE)
```

Update and write the model

```python
model.update()
model.write('Transportation.lp')
```

###5. Model Solution
Solving the model and printing the solution

```python
start = time.time()
model.optimize()
end = time.time()
```

```python
def print_solution(model):
    if model.status != GRB.OPTIMAL:
        print("Model is not Optimized")

    else:
        print('Objective function value: ', model.objVal)
        print('Solution time: ', end - start, ' seconds')
        print('=================================================')
        vars = model.getVars()
        values = model.getAttr('x', vars)
        for var, val in zip(vars, values):
            if val > 1e-6:
                print(f"{var.varName}: {val}")
```

```python
print_solution(model)
```
